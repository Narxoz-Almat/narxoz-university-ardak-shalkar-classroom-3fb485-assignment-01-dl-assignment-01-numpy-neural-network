A.1 Matrix Multiplication (Resource Allocation)

```python
import numpy as np

Resources_Matrix = np.array([
    [10, 20],
    [30, 40],
    [50, 60]
])  # 32

Allocation_Factors = np.array([
    [0.5, 1.0],
    [1.5, 2.0]
])  # 22

Result = Resources_Matrix @ Allocation_Factors
print("Resource Distribution:\n", Result)
```

A.3 Sigmoid Function

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

test_values = np.array([-2, -1, 0, 1, 2])
print("Sigmoid:", sigmoid(test_values))
```

A.4 Sigmoid Gradient

```python
def sigmoid_grad(x):
    s = sigmoid(x)
    return s * (1 - s)

print("Sigmoid Gradient:", sigmoid_grad(test_values))
```

C. NumPy 

```python
import numpy as np

# Dataset
X = np.array([
    [20, 3, 4],
    [15, 5, 3],
    [30, 2, 2],
    [25, 4, 1],
    [35, 2, 3]
]).T  # shape (3,5)

Y = np.array([[18, 20, 22, 25, 30]])

# Initialize
np.random.seed(1)
W1 = np.random.randn(3,3) * 0.01
b1 = np.zeros((3,1))
W2 = np.random.randn(1,3) * 0.01
b2 = np.zeros((1,1))

lr = 0.0001
m = X.shape[1]

def relu(x): return np.maximum(0, x)
def relu_grad(x): return x > 0
def sigmoid(x): return 1/(1+np.exp(-x))

losses = []

for i in range(5000):
    # Forward
    Z1 = W1 @ X + b1
    A1 = relu(Z1)
    Z2 = W2 @ A1 + b2
    A2 = sigmoid(Z2)

    loss = np.mean((A2 - Y)**2)
    losses.append(loss)

    # Backward
    dA2 = 2*(A2 - Y)/m
    dZ2 = dA2 * A2*(1-A2)
    dW2 = dZ2 @ A1.T
    db2 = np.sum(dZ2, axis=1, keepdims=True)

    dA1 = W2.T @ dZ2
    dZ1 = dA1 * relu_grad(Z1)
    dW1 = dZ1 @ X.T
    db1 = np.sum(dZ1, axis=1, keepdims=True)

    # Update
    W2 -= lr*dW2
    b2 -= lr*db2
    W1 -= lr*dW1
    b1 -= lr*db1

print("W1:\n", W1)
print("b1:\n", b1)
print("W2:\n", W2)
print("b2:\n", b2)

# Predictions
Z1 = W1 @ X + b1
A1 = relu(Z1)
Z2 = W2 @ A1 + b2
A2 = sigmoid(Z2)

print("Predictions:\n", A2)
```
